```python
import ollama
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS
import umap
import plotly.graph_objects as go

def ReduceDimensionality(embeddings):
    method = 'UMAP'  # 'PCA', 'UMAP', 'TSNE', or 'MDS'

    X = np.array(embeddings)
        
    if method == 'PCA':
        reducer = PCA(n_components=2)
        coords = reducer.fit_transform(X)        

    elif method == 'UMAP':
        reducer = umap.UMAP(n_components=2, random_state=42)
        coords = reducer.fit_transform(X)
        
        
    elif method == 'TSNE':
        combined = np.vstack([X])
        reduced = TSNE(n_components=2, random_state=42).fit_transform(combined)
        coords = reduced[:len(X)]
                
    elif method == 'MDS':
        combined = np.vstack([X])
        reduced = MDS(n_components=2, random_state=42).fit_transform(combined)
        coords = reduced[:len(X)]
                        
    return coords


def embeddingProcess(embeddingGroups):
        
    items = []  # list of (embedding, sentence, color)

    # Loop through each group of sentences and compute embeddings
    for embeddingGroup in embeddingGroups:        
        sentences = embeddingGroup[0]                        
        color = embeddingGroup[1]
        
        for item in sentences:
            response = ollama.embed( model='nomic-embed-text', input=item)
            items.append((response.embeddings[0], item, color))

    # Reduce dimensionality of embeddings to 2D
    coords = ReduceDimensionality([item[0] for item in items])
        
    fig = go.Figure()
    # Loop through each group of sentences and plot
    for embeddingGroup in embeddingGroups:        
        sentences = embeddingGroup[0]                        
        color = embeddingGroup[1]

        group_indices = [i for i, item in enumerate(items) if item[2] == color]
        group_coords = coords[group_indices]
        group_sentences = [items[i][1] for i in group_indices]
        
        # Add the points to the plot with the specified color and labels
        fig.add_trace(go.Scatter(
            x=group_coords[:, 0],
            y=group_coords[:, 1],
            mode='markers+text', 
            text=group_sentences,
            textposition='top center',
            textfont=dict(size=10),
            marker=dict(size=12, opacity=0.7, color=color),
        ))

    # Set the layout of the plot
    fig.update_layout(
        title=f'Embeddings',
        xaxis_title='Component 1',
        yaxis_title='Component 2',
        width=1000,
        height=750,
    )

    fig.show()
    
```

## 🟢 1 · Array Creation

NumPy arrays (`ndarray`) are faster and more memory-efficient than Python lists because all elements share the same dtype and are stored in contiguous memory.

| Function | Description |
|---|---|
| `np.array()` | From a Python list/tuple |
| `np.zeros/ones/full` | Pre-filled arrays |
| `np.arange` | Like `range()` but returns an array |
| `np.linspace` | N evenly-spaced values between start and stop |
| `np.eye / np.diag` | Identity / diagonal matrix |

> **⚠️ Views vs Copies:** `a[:]` returns a **view** (shares memory). Use `.copy()` to get an independent array.

````python
# From Python sequences
a = np.array([10, 20, 30, 40])
b = np.array([[1, 2, 3], [4, 5, 6]], dtype=float)
print("1D:", a)
print("2D:\n", b)
````

````python
# Built-in constructors
print("zeros(3):        ", np.zeros(3))
print("ones((2,3)):\n",   np.ones((2, 3)))
print("full((2,3),9):\n", np.full((2, 3), 9))
print("eye(3):\n",        np.eye(3, dtype=int))
print("diag([1,2,3]):\n", np.diag([1, 2, 3]))
````

````python
# Ranges & spacing
print("arange(0,10,2): ", np.arange(0, 10, 2))
print("linspace(0,1,6):", np.linspace(0, 1, 6).round(3))
print("logspace(1,3,4):", np.logspace(1, 3, 4).round(2))
````

````python
# Views vs Copies (CRITICAL)
original = np.array([1, 2, 3])
view     = original[:]       # shares memory!
copy     = original.copy()   # independent
original[0] = 99
print("original after change:", original)
print("view  (changed too!): ", view)
print("copy  (unchanged):    ", copy)
print("view.base is original:", view.base is original)
````

## 🟢 2 · Array Attributes & dtypes

Every array carries metadata about its shape and the type of data it holds.

| Attribute | Meaning |
|---|---|
| `.shape` | Tuple of dimension sizes |
| `.ndim` | Number of dimensions |
| `.size` | Total number of elements |
| `.dtype` | Element data type |
| `.itemsize` | Bytes per element |
| `.nbytes` | Total bytes (`size × itemsize`) |

**Choosing the right dtype saves memory.** `float64` uses 8 bytes; `float32` uses 4.

````python
a = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.int32)
print("Array:\n", a)
print("shape    :", a.shape,    "  → (rows, cols)")
print("ndim     :", a.ndim,     "     → number of dimensions")
print("size     :", a.size,     "     → total elements")
print("dtype    :", a.dtype,    " → element type")
print("itemsize :", a.itemsize, "     → bytes per element")
print("nbytes   :", a.nbytes,   "    → total bytes")
````

````python
# Common dtypes
dtypes = [np.int8, np.int16, np.int32, np.int64,
          np.float32, np.float64, np.complex128, np.bool_]
for dt in dtypes:
    arr = np.array([1], dtype=dt)
    print(f"  {str(dt.__name__):15s} → dtype={arr.dtype:12s}  ({arr.itemsize} bytes)")
````

````python
# Type casting with .astype()
x = np.array([1.7, 2.9, 3.1])
print("float64 :", x)
print("→ int   :", x.astype(int),  "  ← truncates, does NOT round!")
print("→ str   :", x.astype(str))
print("round→int:", np.round(x).astype(int))  # round first, then cast
````

## 🟢 3 · Indexing & Slicing

NumPy indexing: `array[start:stop:step]`  
2D: `array[row, col]` — a single comma separates axes.

> Slices return **views**. Negative indices count from the end (`a[-1]` = last element).

````python
a = np.arange(10)
print("array  :", a)
print("a[3]   :", a[3])
print("a[-1]  :", a[-1],   "(last element)")
print("a[2:7] :", a[2:7])
print("a[::2] :", a[::2],  "(every other)")
print("a[::-1]:", a[::-1], "(reversed)")
````

````python
# 2D indexing [row, col]
m = np.arange(1, 13).reshape(3, 4)
print("matrix (3x4):\n", m)
print("m[1, 2]     :", m[1, 2],    "→ row 1, col 2")
print("m[0, :]     :", m[0, :],    "→ entire row 0")
print("m[:, 1]     :", m[:, 1],    "→ entire col 1")
print("m[0:2, 1:3] :\n", m[0:2, 1:3], "→ sub-matrix")
````

````python
# 3D indexing: [layer, row, col]
t = np.arange(24).reshape(2, 3, 4)
print("shape (2,3,4)")
print("t[1, 2, 3]               =", t[1, 2, 3])
print("t[0] (first layer):\n",       t[0])
print("t[:, 1] (row 1 every layer):\n", t[:, 1])
````

## 🟡 4 · Reshaping & Manipulation

Reshaping changes the **view** without copying, as long as total elements stay the same.

| Function | Description |
|---|---|
| `.reshape(shape)` | New shape; `-1` auto-infers that dim |
| `.flatten()` | 1D **copy** |
| `.ravel()` | 1D **view** when possible (faster) |
| `.T` | Transpose |
| `np.insert/delete/append` | Always return **new** arrays — never in-place |

````python
a = np.arange(1, 13)
print("arange(1,13):     ", a)
print("reshape(3,4):\n",    a.reshape(3, 4))
print("reshape(2,2,3):\n", a.reshape(2, 2, 3))
print("reshape(-1,4) → -1 infers rows:\n", a.reshape(-1, 4))
````

````python
m = a.reshape(3, 4)
print("matrix:\n", m)
print("flatten(): ", m.flatten(), " ← always a copy")
print("ravel():   ", m.ravel(),   " ← view (faster)")
print("Transpose (3x4 → 4x3):\n", m.T)
````

````python
# Insert, Delete, Append — all return NEW arrays
x = np.array([1, 2, 3, 4])
print("x:                    ", x)
print("insert(x, 2,[10,11]): ", np.insert(x, 2, [10, 11]))
print("delete(x, 1):         ", np.delete(x, 1))
print("append(x, [5,6]):     ", np.append(x, [5, 6]))
print("x (unchanged):        ", x)  # never modified
````

## 🟡 5 · Math & Statistical Operations

Statistics can be computed over the **whole array** or along a specific **axis**.

```
axis=0  →  collapse rows   (result per column)
axis=1  →  collapse columns (result per row)
```

**Normalization** is essential in ML preprocessing:
- **Min-Max** scales to `[0, 1]`
- **Z-score** centers around 0 with std = 1

````python
data = np.array([4., 7., 13., 2., 1., 9., 15., 6., 11., 8.])
print("data       :", data)
print("sum        :", data.sum())
print("mean       :", data.mean())
print("median     :", np.median(data))
print("std        :", round(data.std(), 4))
print("var        :", round(data.var(), 4))
print("min / max  :", data.min(), "/", data.max())
print("argmin/max :", data.argmin(), "/", data.argmax(), " ← indices")
print("percentiles [25,50,75]:", np.percentile(data, [25, 50, 75]))
````

````python
# Axis-aware operations
m = np.array([[1, 2, 3],
              [4, 5, 6]])
print("matrix:\n", m)
print("sum(axis=0): ", m.sum(axis=0),  " ← collapse rows  (per column)")
print("sum(axis=1): ", m.sum(axis=1),  " ← collapse cols  (per row)")
print("mean(axis=0):", m.mean(axis=0))
print("cumsum():    ", m.cumsum(),      " ← running total (flattened)")
````

````python
# Math functions
x = np.array([0., np.pi/6, np.pi/4, np.pi/2])
print("sin(x)        :", np.sin(x).round(3))
print("cos(x)        :", np.cos(x).round(3))
print("exp([0,1,2])  :", np.exp([0, 1, 2]).round(3))
print("log([1,e,e^2]):", np.log([1, np.e, np.e**2]).round(3))
print("sqrt([4,9,16]):", np.sqrt([4, 9, 16]))
print("abs([-3,4,-1]):", np.abs([-3, 4, -1]))
print("power([2,3],3):", np.power([2, 3], 3))
````

````python
# Normalization
d = np.array([10., 20., 30., 40., 50.])
min_max = (d - d.min()) / (d.max() - d.min())
z_score = (d - d.mean()) / d.std()
print("data          :", d)
print("min-max [0,1] :", min_max)
print("z-score       :", z_score.round(2))
````

````python
# Clipping & rounding
noisy = np.array([5., 15., 150., -10., 30.])
print("data       :", noisy)
print("clip(0,50) :", np.clip(noisy, 0, 50))

vals = np.array([-1.7, 0.5, 2.3, 4.8])
print("round(vals):", np.round(vals))
print("floor(vals):", np.floor(vals))
print("ceil(vals) :", np.ceil(vals))
````

## 🟡 6 · Boolean & Fancy Indexing

Boolean indexing **filters** elements by condition. Fancy indexing selects by an **array of indices**.

> ⚠️ Use `&` `|` `~` for AND / OR / NOT — **not** Python's `and`/`or`/`not`.

| Tool | Use case |
|---|---|
| `a[mask]` | Filter by boolean array |
| `np.where(cond, x, y)` | Ternary replacement |
| `a[idx_array]` | Select by integer indices |
| `np.select(conds, choices)` | Multi-condition branching |

````python
a = np.array([3, 8, 15, 22, 30, 41, 5, 18])
mask = a > 15
print("array            :", a)
print("mask (>15)       :", mask)
print("a[mask]          :", a[mask])
print("a[a%2==0]        :", a[a % 2 == 0],       " ← even numbers")
print("a[(a>5)&(a<25)]  :", a[(a > 5) & (a < 25)])
````

````python
# np.where — indices OR ternary replacement
print("array                 :", a)
print("where(a>15) indices   :", np.where(a > 15)[0])
replaced = np.where(a > 20, 99, a)
print("where(a>20, 99, a)    :", replaced, " ← ternary replacement")
````

````python
# Fancy indexing — integer array as index
b   = np.array([100, 200, 300, 400, 500])
idx = np.array([0, 2, 4, 2])
print("array  :", b)
print("idx    :", idx)
print("b[idx] :", b[idx], " ← repeats allowed")
````

````python
# np.select — multi-condition branching
grades  = np.array([45, 72, 88, 55, 95, 61])
conds   = [grades >= 90, grades >= 75, grades >= 60]
choices = ["A",           "B",          "C"]
labels  = np.select(conds, choices, default="F")
print("grades :", grades)
print("labels :", labels)
````

## 🟡 7 · Broadcasting

Broadcasting auto-expands smaller arrays to match larger ones **without copying memory**.

**Rules (align dims from the RIGHT):**
1. Dimensions must be equal, or
2. One of them must be `1`

```
(3,)   →  (1, 3)  compatible with (4, 3)
(3, 1)             compatible with (3, 4)
```

`np.newaxis` inserts a size-1 dimension to enable broadcasting.

````python
# Scalar broadcast
a = np.array([1, 2, 3])
print("[1,2,3] + 10 =", a + 10)
print("[1,2,3] * 2  =", a * 2)
````

````python
# Vector broadcast over matrix
m   = np.ones((3, 3), dtype=int)
row = np.array([10, 20, 30])        # shape (3,)
col = np.array([[1], [2], [3]])     # shape (3,1)
print("matrix:\n", m)
print("+ row vector [10,20,30]:\n",   m + row, " ← adds to each row")
print("+ col vector [[1],[2],[3]]:\n", m + col, " ← adds to each col")
````

````python
# np.newaxis — outer product
r = np.arange(1, 4)   # shape (3,)
c = np.arange(1, 5)   # shape (4,)
outer = r[:, np.newaxis] * c   # (3,1) * (4,) → (3,4)
print(f"r={r}  x  c={c}  →  outer product:")
print(outer)
````
